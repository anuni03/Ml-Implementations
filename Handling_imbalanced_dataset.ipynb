**Imbalanced** **Dataset**

A dataset with an unequal class distribution

```python
# importing the dependencies
import numpy as np
import pandas as pd
```

```python
#loading the dataset to pandas Dataframe
credit_card_data=pd.read_csv('/credit_data.csv')
```

```python
# first five rows of the dataframe
credit_card_data.head()
```

```python
# distribution of the two classes
credit_card_data['Class'].value_counts()
```

This is highly imbalanced dataset

0 ----> Legit Transactions
1 -----> Fraudulent Transactions

```python
#seperating the legit and fraudulent transactions

legit=credit_card_data[credit_card_data.Class ==0]
fraud=credit_card_data[credit_card_data.Class ==1]
```

```python
print(legit.shape)
print(fraud.shape)
```

Under - sampling

build a sample dataset containing similiar distribution of Legit and Fraudent Transactions

```python
legit_same=legit.sample(n=148)
```

```python
print(legit_same.shape)
```

Concatenate the two DataFrames

```python
new_dataset=pd.concat([legit_same,fraud],axis=0)
```
